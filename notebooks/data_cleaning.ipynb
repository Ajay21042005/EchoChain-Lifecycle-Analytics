```python
import pandas as pd
```

```python
olx = pd.read_excel("../data/raw/olx_dataset.xlsx")

ebay = pd.read_csv(
    "../data/raw/ebay_products.csv",
    engine="python",
    on_bad_lines="skip"
)

used = pd.read_csv("../data/raw/used_devices.csv")
```

```python
print("OLX duplicates:", olx.duplicated().sum())
print("eBay duplicates:", ebay.duplicated().sum())
print("Used duplicates:", used.duplicated().sum())
```

```python
olx = olx.drop_duplicates()
ebay = ebay.drop_duplicates()
used = used.drop_duplicates()
print("Duplicates removed!")
```

```python
print("OLX")
print(olx.isnull().sum())
print("\neBay")
print(ebay.isnull().sum())
print("\nUsed Devices")
print(used.isnull().sum())
```

```python
olx = olx.fillna(olx.median(numeric_only=True))
ebay = ebay.fillna(ebay.median(numeric_only=True))
used = used.fillna(used.median(numeric_only=True))
```

```python
olx = olx.fillna("Unknown")
ebay = ebay.fillna("Unknown")
used = used.fillna("Unknown")
```

```python
print(olx.dtypes)
print(ebay.dtypes)
print(used.dtypes)
```

```python
olx.to_csv("../data/processed/olx_cleaned.csv", index=False)
ebay.to_csv("../data/processed/ebay_cleaned.csv", index=False)
used.to_csv("../data/processed/used_devices_cleaned.csv", index=False)
print("Cleaned datasets saved successfully!")
```
